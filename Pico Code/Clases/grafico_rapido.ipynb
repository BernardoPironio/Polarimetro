```python
import serial
import time

ser = serial.Serial("COM7",115200,timeout=1)

time.sleep(2)

for _ in range(100):
    ser.readline()

while True:
    linea = ser.readline().decode(errors="ignore").strip()

    if linea:
        print(linea)
```
