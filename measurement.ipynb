# Oscilloscope connection

```python
rm = pyvisa.ResourceManager()
display(rm.list_resources())
osci = labo_instruments.TDS1002B('USB0::0x0699::0x0368::C033877::INSTR')
```

# Manual voltage measurment

```python
# --- Gets the first channel configuration values of the oscilloscope ---
osci.get_channel(1)
```

```python
# --- Creates an ampty list to store the voltage values ---
voltaje = []
```

```python
# --- Adds the value of the voltaje every time you run the cell, and displays the list and its length ---
voltaje.append(osci.read_data(1)[1].mean())
display(voltaje)
display(len(voltaje))
```

```python
angulo = np.arange(0,361,10)
angulo[0:-1]
```

```python
# --- Converts the list of voltaje to a dataframe and saves it as a csv file ---
data = pd.DataFrame({'voltaje':voltaje,'angulo':angulo[0:-1]}).to_csv('Measurements/medicion_manual_polarimetro_ortogonal.csv', index=False)
```
